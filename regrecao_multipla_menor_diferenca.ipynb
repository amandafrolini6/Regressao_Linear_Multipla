```python
import numpy as np
import pandas as pd

# Tente achar a menor diferença de preço

# 1. Lendo os dados com NumPy
# Pulamos o cabeçalho (skip_header=1)
# Extraimos apenas as colunas de interesse para evitar erro com string:
# 2: price (Y), 3: bedrooms (X1), 4: bathrooms (X2), 5: condition (X3)
dados = np.genfromtxt('kc_house_data.csv', delimiter=',', skip_header=1, usecols=(2, 3, 4, 10))
df_dados = pd.read_csv('kc_house_data.csv')
# 2. Separando a variável alvo (Y) e as variáveis preditoras (X)
y = dados[:, 0] # O preço é a primeira coluna do nosso recorte
X_features = dados[:, 1:] # O restante são as features 

# 3. Adicionando o termo de Intercepto (Viés/Bias)
# Precisamos adicionar uma coluna de números "1" no início da matriz X
# Isso represanta o beta_0 na equação (o valor base da casa se tudo fosse zero)
m = X_features.shape[0] # Número de linhas (exemplo)
X = np.c_[np.ones(m), X_features]

# 4. Aplicando a Equação Normal: Beta = (X.T * X)^-1 * X.T * Y
X_T = X.T # Transposta de X
# np.linealg.inv calcula a matriz inversa da matriz
# O método .dot() ou o operador @ realiza a multiplicação de matrizes
beta = np.linalg.inv(X_T @ X) @ X_T @ y

print("=== Coeficientes Descobertos ===")
print(f"Intercepto (Beta 0): {beta[0]:.2f}")
print(f"Peso para Quartos (Beta 1): {beta[1]:.2f}")
print(f"Peso para Banheiros (Beta 2): {beta[2]:.2f}")
print(f"Peso para a Condição (Beta 3): {beta[3]:.2f}\n")

df_dados.info()
# 5. Fazendo uma previsão 
# Vamos prever o valor da primeira casa do dataset: 3 quartos, 1 banheiro, 3 condição
casa_teste = np.array([1, 3, 1, 3]) # Lembre-se do "1" inicial para o intercepto!
preco_previsto = casa_teste @ beta

print("\n=== Teste de Previsão ===")
print(f"Preço REAL da casa 1: $ {y[0]:.2f}")
print(f"Preço PREVISTO pelo nosso modelo: $ {preco_previsto:.2f}")
```
